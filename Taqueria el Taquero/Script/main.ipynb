# Taqueria el Taquero

```python
import pandas as pd
from pathlib import Path

# Ruta hacia la carpeta Data
data_dir = Path("../Data")

if not data_dir.exists():
    raise FileNotFoundError(f"No se encontro la carpeta Data en: {data_dir} :(")

print("Carpeta Data encontrada:", data_dir)
```

```python
# Orden personalizado de los archivos CSV
orden_csv = [
    "Cliente.csv",
    "Empleado.csv",
    "Proveedor.csv",
    "Producto.csv",
    "Inventario.csv",
    "Pedido.csv",
    "DetallePedido.csv",
    "MetodoPago.csv",
    "Pago.csv",
    "Evento.csv"
]

for archivo in orden_csv:
    ruta = data_dir / archivo
    print("\n---")
    print(f"Archivo: {archivo}")

    if not ruta.exists():
        print("No se encontro este archivo dentro de /Data :(")
        continue

    df = pd.read_csv(ruta)
    display(df.head(30))
    print(f"Filas: {len(df)}, Columnas: {len(df.columns)}")
```
